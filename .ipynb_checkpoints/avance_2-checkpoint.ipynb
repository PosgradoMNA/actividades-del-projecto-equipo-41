## CRISP - DM 
> La metodología a utilizar será **CR**oss-**I**ndustry **S**tandard **P**rocess for **D**ata **M**ining (CRISP-DM), la cual consta de los siguientes pasos:
> 
> 1. Entendimiento del negocios
> <br> Se plantea un problema inicial, sus objetivos y requerimentos. En este paso también se evalúa la situación actual y se plantean los objetivos esperados.
> 2. Entendimiento de los datos
> <br> Se recolectan los datos iniciales, se ve la calidad de estos y se hace un análisis exploratorio que puede sugerir hipótesis más adelante,
> 3. Preparación de los datos
> <br> En este paso ponemos los datos en estándares entendibles para los algoritmos, limpiamos errores, y nos aseguramos de tener casos completos. En este paso podemos además integrar otras fuentes que nos sirvan de apoyo.
> 4. Modelado
> <br> Se determinan algortimos a utilizar para entender los datos, como pueden ser: regresiones lineales, redes neuronales, árboles de decisión, etc. En este paso, se aplican y ajustan los hiperparemetros de los modelos para alcanzar el objetivo.
> 5. Evaluación 
> <br> Se determina si el ejercicio cumplió con las expectativas. En este paso puede ocurrir un ciclo de vuelta al entendimiento del negocio, pues probablemente se encuentren limitantes o nuevas posibilidades que quieran ser incluidas en el análisis.
> 6. Implementación 
> <br> Cuando la evaluación es satisfacrtoria para los objetivos se pone en práctica o producción los modelos para alcanzar los resultados. Las buenas practicas nos indican que aquí también se hace la documentación del proceso y se puede plantear nuevos escenarios futuros.
